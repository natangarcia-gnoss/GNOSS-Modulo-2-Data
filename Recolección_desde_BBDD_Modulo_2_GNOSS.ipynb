Con el siguiente codigo podemos obtener el porcentaje de campos

```python
total = df.size
for col in df.columns:
  df_fil = df[df[col].notna()]
  size= df_fil.size
  percent = (df_fil.size/total)*100
  print("Field " + col + " has " + str(size) + " values of " + str(total) + " (" + str(percent) + "%)")
```

Procedemos a eliminar las columnas que tienen pocos datos puesto que no nos aportan informacion significativa (Supongo un umbral arbitrario de por debajo del 70% de registros)

```python
df.columns
```

```python
cols_to_delete = []
total = df.size
for col in df.columns:
  df_fil = df[df[col].notna()]
  size= df_fil.size
  if (df_fil.size/total) < 0.7:
    cols_to_delete.append(col)

df = df.drop(columns=cols_to_delete)
print("Se han eliminado las columnas: " + ', '.join(cols_to_delete))
```
